```python
import pandas as pd
```

```python
df = pd.read_json('watcher_crawl/watcher.jl', lines=True)
```

```python
df.head(5)
```

```python
df['table_name'] = df['company_code'].astype(str) + '-' + df['scraped_table'].astype(str)
df
```

```python
table_names = list(df['table_name'])
print(table_names)
```

```python
import pymongo
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
m_db = myclient["stock_database_final"]
```

```python
dblist = myclient.list_database_names()
if "mydatabase" in dblist:
  print("The database exists.")
```

```python
for i in range(df.shape[0]):
    print("\nTable Name ------------"+table_names[i])
    df2 = pd.DataFrame(df['records'][i].get('values'), columns = df['records'][i].get('columns')).T
    print(df['records'][i].get('columns'))
    print(df2)
```

```python
for i in range(df.shape[0]):
    table_name = m_db[table_names[i]]
    column_names = df['records'][i].get('columns')
    df2 = pd.DataFrame(df['records'][i].get('values'), columns = column_names)
    #df2.index = [str(i) for i in range(len(df2))]
    df2.set_index('fields',inplace=True)
    df3 = df2.T
    table_name.insert_many(df3.to_dict('records'))
    for i in table_name.find():
        print(i)

```
